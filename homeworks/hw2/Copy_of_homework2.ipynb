## Data Cleaning

````python
import json
import re

# Data Cleaning: Convert resume text to structured JSON format using LLM

clean_cv_prompt = ChatPromptTemplate.from_template("""
Extract and structure the following resume into a clean JSON format.

Extract the following fields if present:
- name: Candidate's full name
- email: Email address
- phone: Phone number
- location: Location/City
- summary: Brief professional summary (1-2 sentences)
- education: List of educational background
- experience: List of work experience with job title, company, duration, and key responsibilities
- skills: List of key skills
- certifications: Any relevant certifications

Resume text:
{cv_text}

Output ONLY a valid JSON object (no markdown, no explanation).
""")

cleaning_chain = clean_cv_prompt | llm | StrOutputParser()

def clean_json_string(json_str):
    """Remove markdown code block markers and extract pure JSON"""
    # Remove markdown code block markers ```json and ```
    json_str = re.sub(r'^```json\s*', '', json_str, flags=re.MULTILINE)
    json_str = re.sub(r'^```\s*$', '', json_str, flags=re.MULTILINE)
    json_str = json_str.strip()
    return json_str

# Clean each resume
cleaned_cvs = []

print("=" * 80)
print("ğŸ§¹ Starting resume data cleaning...")
print("=" * 80)

for i, cv in enumerate(all_cvs):
    try:
        # Call LLM to clean the resume
        cleaned_json_str = await cleaning_chain.ainvoke({"cv_text": cv['text']})

        # Clean markdown markers
        cleaned_json_str = clean_json_string(cleaned_json_str)

        # Try to parse JSON (validate)
        try:
            cleaned_data = json.loads(cleaned_json_str)
            cleaned_cvs.append({
                "file": cv['file'],
                "original_text": cv['text'],
                "cleaned_json": cleaned_json_str,
                "parsed_data": cleaned_data
            })
            print(f"âœ“ Cleaning completed: {cv['file']}")
            print(f"  Key fields: name={cleaned_data.get('name', 'N/A')}, email={cleaned_data.get('email', 'N/A')}, location={cleaned_data.get('location', 'N/A')}")
        except json.JSONDecodeError as e:
            print(f"âš ï¸ {cv['file']} JSON parsing failed: {e}")
            print(f"  Raw output: {cleaned_json_str[:100]}...")
            cleaned_cvs.append({
                "file": cv['file'],
                "original_text": cv['text'],
                "cleaned_json": cleaned_json_str,
                "parsed_data": None
            })
    except Exception as e:
        print(f"âš ï¸ Cleaning failed: {cv['file']} - {e}")
        cleaned_cvs.append({
            "file": cv['file'],
            "original_text": cv['text'],
            "cleaned_json": "",
            "parsed_data": None
        })

print(f"\nâœ“ Cleaning completed, processed {len(cleaned_cvs)} resumes")
print("=" * 80)
````

## Info Fetching

````python
import asyncio
import json
from datetime import datetime

info_fetched_results = []

system_prompt = """You are a Resume Information Fetcher. Your task is to:
1. Use LinkedIn and Facebook tools to search for the candidate
2. Retrieve their profiles from both platforms
3. Extract and return structured information in JSON format

Workflow:
1. Search LinkedIn using candidate's name and location
2. Get the best matching LinkedIn profile (pick most relevant match)
3. Search Facebook using candidate's name
4. Get 1-2 Facebook profiles if found
5. **After retrieving profiles, STOP and output JSON**

Output format (after gathering data):
```json
{
  "linkedin_profile": {
    "id": ...,
    "name": "...",
    "headline": "...",
    "location": "...",
    "industry": "...",
    "experience": [...],
    "education": [...]
  },
  "facebook_profiles": [
    {
      "id": ...,
      "name": "...",
      "location": "...",
      "bio": "..."
    }
  ],
  "search_summary": "Brief summary of what was found"
}
```

**Important**: After 3-5 tool calls, you MUST stop and output the final JSON."""

async def process_cv_with_tools(cv_info, tools):
    """Process a single CV by calling LLM and invoking tools as needed."""
    file_name = cv_info['file']
    cleaned_data = cv_info.get('parsed_data') or cv_info.get('cleaned_data')

    print(f"\nğŸ”„ Processing {file_name}...")

    user_content = f"""Candidate Information:
{json.dumps(cleaned_data, ensure_ascii=False, indent=2)}

Please use LinkedIn and Facebook tools to search for this candidate and retrieve their profiles.
After gathering the information, output a JSON summary of what you found."""

    messages = [
        ("system", system_prompt),
        ("human", user_content)
    ]

    max_iterations = 6
    iteration = 0
    tool_count = 0

    while iteration < max_iterations:
        iteration += 1
        print(f"  Iteration {iteration}...")

        try:
            response = llm_with_tools.invoke(messages)

            if hasattr(response, 'tool_calls') and response.tool_calls:
                messages.append(response)

                for tool_call in response.tool_calls:
                    if isinstance(tool_call, dict):
                        tool_name = tool_call.get('name')
                        tool_input = tool_call.get('args', {})
                        tool_id = tool_call.get('id', str(iteration))
                    else:
                        tool_name = getattr(tool_call, 'name', None)
                        tool_input = getattr(tool_call, 'args', {}) or {}
                        tool_id = getattr(tool_call, 'id', str(iteration))

                    if not tool_name:
                        continue

                    print(f"    ğŸ“ Calling tool: {tool_name}")
                    tool_count += 1

                    tool_found = False
                    for tool in tools:
                        if tool.name == tool_name:
                            tool_found = True
                            try:
                                tool_result = await tool.ainvoke(tool_input)
                                tool_result_str = json.dumps(tool_result) if isinstance(tool_result, dict) else str(tool_result)

                                messages.append(ToolMessage(content=tool_result_str, tool_call_id=tool_id))

                                print(f"    âœ… Tool result received ({len(tool_result_str)} chars)")
                            except Exception as e:
                                print(f"    âš ï¸ Tool invocation error: {str(e)[:100]}")
                                messages.append(ToolMessage(content=f"Error: {str(e)}", tool_call_id=tool_id))
                            break

                    if not tool_found:
                        messages.append(ToolMessage(
                            content=f"Tool {tool_name} not found.",
                            tool_call_id=tool_id
                        ))

                if iteration >= 4:
                    messages.append(("human", "You have enough information now. Please provide your final JSON summary and stop using tools."))

            else:
                final_response = response.content if hasattr(response, 'content') else str(response)
                print(f"  âœ… Processing complete")

                clean_response = final_response.strip()
                if clean_response.startswith('```'):
                    clean_response = clean_response.split('```')[1]
                    if clean_response.startswith('json'):
                        clean_response = clean_response[4:]
                    clean_response = clean_response.split('```')[0]
                clean_response = clean_response.strip()

                try:
                    fetched_data = json.loads(clean_response)
                except:
                    fetched_data = {"raw_response": final_response}

                info_fetched_results.append({
                    "file": file_name,
                    "cleaned_data": cleaned_data,
                    "fetched_data": fetched_data,
                    "tool_count": tool_count
                })
                return True

        except Exception as e:
            print(f"  âŒ Error in iteration {iteration}: {str(e)[:150]}")
            break

    print(f"  â±ï¸ Max iterations reached, requesting final summary...")
    try:
        messages.append(("human", "Please provide your final JSON summary NOW based on the information you gathered."))
        final_response = llm_with_tools.invoke(messages)
        final_text = final_response.content if hasattr(final_response, 'content') else str(final_response)

        clean_text = final_text.strip()
        if clean_text.startswith('```'):
            clean_text = clean_text.split('```')[1]
            if clean_text.startswith('json'):
                clean_text = clean_text[4:]
            clean_text = clean_text.split('```')[0]
        clean_text = clean_text.strip()

        try:
            fetched_data = json.loads(clean_text)
        except:
            fetched_data = {"raw_response": final_text, "incomplete": True}
    except:
        fetched_data = {"error": "Failed to get summary", "incomplete": True}

    info_fetched_results.append({
        "file": file_name,
        "cleaned_data": cleaned_data,
        "fetched_data": fetched_data,
        "tool_count": tool_count
    })
    return False

# Process all CVs
print(f"Starting to process {len(cleaned_cvs)} resumes with tools...")
print(f"Available tools: {[t.name for t in mcp_tools]}")

try:
    # Run async processing for all CVs
    results = []
    for cv in cleaned_cvs:
        success = await process_cv_with_tools(cv, mcp_tools)
        results.append(success)

    successful = sum(results)
    print(f"\n{'='*50}")
    print(f"âœ… Processing complete: {successful}/{len(cleaned_cvs)} successful")
    print(f"{'='*50}")
except Exception as e:
    print(f"âŒ Unexpected error: {e}")
    import traceback
    traceback.print_exc()
````

## Rating

````python
import re


scoring_prompt = ChatPromptTemplate.from_template("""You are an expert at comparing resume data for consistency and authenticity verification.

Given the following two data sources:
1. CLEANED_DATA: Extracted from the original CV document
2. FETCHED_DATA: Retrieved from LinkedIn and Facebook profiles

Your task is to evaluate the CONSISTENCY between these two sources.
Consider:
- Name matching (including variations, middle names, initials)
- Location/geography consistency
- Education background alignment
- Work experience alignment (job titles, companies, timeline)
- Skills overlap
- Overall coherence and plausibility

IMPORTANT:
- Minor discrepancies (typos, abbreviations, order differences) are normal and acceptable
- Major contradictions (different names, completely different companies, timeline conflicts) indicate lower consistency
- Missing information in one source is acceptable (doesn't lower score much)
- Suspicious patterns (fabricated profiles, clearly fake data) significantly lower score

Output ONLY a JSON object with:
{{
  "consistency_score": <float between 0.0 and 1.0>,
  "reasoning": "<brief explanation of the score>",
  "matches": [<list of verified matching fields>],
  "discrepancies": [<list of notable differences if any>]
}}

CLEANED_DATA:
{cleaned_data}

FETCHED_DATA:
{fetched_data}

Respond with ONLY the JSON object, no markdown or additional text.""")

async def semantic_score_cv(cleaned_data, fetched_data):
    """
    Use LLM to semantically compare cleaned_data and fetched_data
    Returns a score between 0.0 and 1.0
    """
    scoring_chain = scoring_prompt | llm | StrOutputParser()

    try:
        response = await scoring_chain.ainvoke({
            "cleaned_data": json.dumps(cleaned_data, ensure_ascii=False, indent=2),
            "fetched_data": json.dumps(fetched_data, ensure_ascii=False, indent=2)
        })

        # Clean markdown if present
        clean_response = response.strip()
        if clean_response.startswith('```'):
            clean_response = clean_response.split('```')[1]
            if clean_response.startswith('json'):
                clean_response = clean_response[4:]
            clean_response = clean_response.split('```')[0]
        clean_response = clean_response.strip()

        result = json.loads(clean_response)
        score = float(result.get('consistency_score', 0.5))

        # Clamp score to [0.0, 1.0]
        score = max(0.0, min(1.0, score))

        return {
            "score": score,
            "reasoning": result.get('reasoning', ''),
            "matches": result.get('matches', []),
            "discrepancies": result.get('discrepancies', [])
        }
    except Exception as e:
        print(f"âš ï¸ Scoring error: {str(e)[:100]}")
        return {
            "score": 0.5,
            "reasoning": f"Scoring error: {str(e)}",
            "matches": [],
            "discrepancies": []
        }

# Run semantic scoring on all fetched results
print("=" * 80)
print("ğŸ¤– Semantic Scoring Phase (LLM-based comparison)")
print("=" * 80)

final_scores = []

try:
    for i, result in enumerate(info_fetched_results):
        file_name = result['file']
        cleaned_data = result['cleaned_data']
        fetched_data = result['fetched_data']

        print(f"\nğŸ“Š Scoring {file_name}...")

        if cleaned_data is None:
            print(f"  âš ï¸ No cleaned data available, score = 0.0")
            final_scores.append(0.0)
            continue

        if not fetched_data or 'error' in fetched_data:
            print(f"  âš ï¸ No valid fetched data, score = 0.0")
            final_scores.append(0.0)
            continue

        score_result = await semantic_score_cv(cleaned_data, fetched_data)
        score = score_result['score']

        final_scores.append(score)

        print(f"  ğŸ“ˆ Score: {score:.2f}")
        print(f"  ğŸ’­ Reasoning: {score_result['reasoning']}")
        if score_result['matches']:
            print(f"  âœ… Matches: {', '.join(score_result['matches'][:3])}")
        if score_result['discrepancies']:
            print(f"  âš ï¸ Discrepancies: {', '.join(score_result['discrepancies'][:3])}")

    print(f"\n{'='*80}")
    print(f"âœ… Scoring complete!")
    print(f"Final scores: {[f'{s:.2f}' for s in final_scores]}")
    print(f"{'='*80}")

except Exception as e:
    print(f"âŒ Unexpected error during scoring: {e}")
    import traceback
    traceback.print_exc()
````

# Evaluation code

In the test phase, you will be given 5 CV files with fixed names:

    CV_1.pdf, CV_2.pdf, CV_3.pdf, CV_4.pdf, CV_5.pdf

Your system must process these CVs and output a list of 5 scores,
one score per CV, in the same order:

    scores = [s1, s2, s3, s4, s5]

Each score must be a float in the range [0, 1], representing the
reliability or confidence that the CV is valid (or meets the task criteria).

The ground-truth labels are binary:

    groundtruth = [0 or 1, ..., 0 or 1]

Each CV is evaluated independently using a threshold of 0.5:

- If score > 0.5 and groundtruth == 1 â†’ Full credit
- If score â‰¤ 0.5 and groundtruth == 0 â†’ Full credit
- Otherwise â†’ No credit

In other words, 0.5 is the decision threshold.

- Each CV contributes equally.
- Final score = (number of correct decisions) / 5


````python
# =====================================================
#  Evaluation code
# =====================================================

def evaluate(scores, groundtruth, threshold=0.5):
    """
    scores: list of floats in [0, 1], length = 5
    groundtruth: list of ints (0 or 1), length = 5
    """
    assert len(scores) == 5
    assert len(groundtruth) == 5

    correct = 0
    decisions = []

    for s, gt in zip(scores, groundtruth):
        pred = 1 if s > threshold else 0
        decisions.append(pred)
        if pred == gt:
            correct += 1

    final_score = correct / len(scores)

    return {
        "decisions": decisions,
        "correct": correct,
        "total": len(scores),
        "final_score": final_score
    }
````
